```python
df.drop('month',axis=1,inplace=True)
df.drop('year',axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder

df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['weather']=le.fit_transform(df['weather'])
```

```python
x = df[['temp_min', 'temp_max', 'precipitation', 'wind']]
y=df['weather']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

nb_model = GaussianNB()
nb_model.fit(X_train, y_train)

y_pred = nb_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred, zero_division=1)


print(f"Accuracy: {accuracy:.2f}")
```

```python
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(classification_rep)
```

```python
# Get user inputs
temp_min_input = float(input("Enter minimum temperature (°C): "))
temp_max_input = float(input("Enter maximum temperature (°C): "))
precipitation_input = float(input("Enter precipitation (mm): "))
wind_input = float(input("Enter wind speed (km/h): "))

# Create a DataFrame for the input
user_input = pd.DataFrame([[temp_min_input, temp_max_input, precipitation_input, wind_input]],
                          columns=['temp_min', 'temp_max', 'precipitation', 'wind'])

# Predict the weather using the trained model
predicted_weather_encoded = nb_model.predict(user_input)

# Reverse the encoding to get the original weather label
predicted_weather = le.inverse_transform(predicted_weather_encoded)

print(f"\nPredicted weather: {predicted_weather[0]}")
```
